```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dat = pd.read_csv("https://raw.githubusercontent.com/rerkki/metropolia/master/waveforms.csv")
dat[0:2] #tulostetaan otsikko ja ensimmäinen rivi tarkistuksen vuoksi
```

```
t = dat['t']
s = dat['s6'] # Muuta tässä kohdassa sulkeiden sisään s1 ... s7 vaihtaaksesi aaltomuotoa
plt.plot(s)
plt.show()
```

```
fft = np.fft.fft(s)
T = t[1] - t[0]  # näytteenottoväli 
N = s.size

f = np.linspace(0, 1/(T), N)

plt.ylabel("Amplitudi")
plt.xlabel("Taajuus [Hz]")
plt.plot(f[:N // 2], np.abs(fft)[:N//2]*1/N)
plt.show()
```
