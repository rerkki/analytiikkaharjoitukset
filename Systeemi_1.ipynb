```
import pandas as pd
import matplotlib.pyplot as plt
```

Haetaan data ja tulostetaan siitä otsikkorivi ja ensimmäinen datarivi näkyviin

```
data = pd.read_csv("https://raw.githubusercontent.com/rerkki/metropolia/master/papermachine.csv", encoding='latin-1')
data[0:1]
```

```
#määritellään muuttujat datasarakkeille:
gr = data['WW_pH'] 
ret = data['WW_CONDUCTIVITY']

#määritellään kuvan otsikot:
plt.ylabel("COND")
plt.xlabel("Datapoint")

#piirretään datat kuvaajaan:
#plt.plot(gr)
plt.plot(ret)

#näytetään kuva:
plt.show()
```

```
#määritellään muuttujat datasarakkeille:
gr = data['GRAMMAGE'] 
ret = data['RETENTION_RMI']

#piirretään kuva toisin, eli neliöpaino vs. retentio:
plt.ylabel("RETENTION")
plt.xlabel("GRAMMAGE")
plt.scatter(gr, ret)
plt.show()
#tässä kuvassa nähdään korrelaatio neliöpainon ja retention välillä
#nähdään myös voimakas prosessihäiriö
```
