```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
data_ = pd.read_csv("https://raw.githubusercontent.com/rerkki/metropolia/master/papermachine.csv", encoding='latin-1')
data = data_.drop(data_.index[1620]) #datan siivousta, poistetaan yksi rivi
data_normal = data.drop(data.index[2100:2380]) #jaetaan data normaalitilanteeseen...
data_problem = data.loc[2100:2300] # ...ja ongelmatilanteeseen

data[0:1] #tulostetaan otsikko ja ensimmäinen rivi tarkistuksen vuoksi
```

```
fig, ax = plt.subplots(3, figsize = (10,10))
#tulostetaan kuvaajat jotta nähdään että datan jakaminen osajoukkoihin onnistui
ax[0].plot(data['RETENTION_RMI']) #kaikki data
ax[1].plot(data_normal['RETENTION_RMI']) #datasta poistettu ongelmatilanne
ax[2].plot(data_problem['RETENTION_RMI']) #ongelmatilanteen data pelkästään
```

```
# Valitse datan osajoukko

dat = data #kaikki data
#dat = data_normal #normaalitilanteen data
#dat = data_problem #ongelmatilanteen data
```

```
# Tehdään korrelaatiokartta kaikista muuttujista valitsemallesi datan osajoukolle
corr = dat.corr()

mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
fig, ax = plt.subplots(figsize=(18, 10)) 
sns.heatmap(corr, mask=mask, cmap=sns.diverging_palette(256,0,sep=80,n=7,as_cmap=True));
```
